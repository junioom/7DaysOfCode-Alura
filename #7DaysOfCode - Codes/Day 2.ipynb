# DIA 1

## Importando as bases de dados dos emprestimos e concatenando todas.

```python
import pandas as pd

base_url = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/'

files = ['emprestimos-20101.csv',
         'emprestimos-20102.csv',
         'emprestimos-20111.csv',
         'emprestimos-20112.csv',
         'emprestimos-20121.csv',
         'emprestimos-20122.csv',
         'emprestimos-20131.csv',
         'emprestimos-20132.csv',
         'emprestimos-20141.csv',
         'emprestimos-20142.csv',
         'emprestimos-20151.csv',
         'emprestimos-20152.csv',
         'emprestimos-20161.csv',
         'emprestimos-20162.csv',
         'emprestimos-20171.csv',
         'emprestimos-20172.csv',
         'emprestimos-20181.csv',
         'emprestimos-20182.csv',
         'emprestimos-20191.csv',
         'emprestimos-20192.csv',
         'emprestimos-20201.csv']


url_file_emprestimo = [base_url + file for file in files]

dfs_emprestimos = []

for url in url_file_emprestimo:
    df = pd.read_csv(url)
    dfs_emprestimos.append(df)

df_combinado_emprestimo = pd.concat(dfs_emprestimos, ignore_index=True)
```

## Verificando se há duplicatas (Informações duplicadas)

```python
df_combinado_emprestimo.value_counts()
```

### Removendo-as:

```python
df_combinado_emprestimo = df_combinado_emprestimo.drop_duplicates()

df_combinado_emprestimo.value_counts()
```

## Importando a segunda base de dados (Exemplares)

```python
database_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
display(database_exemplares)
```

## Mesclando as duas bases de dados

```python
df_final = df_combinado_emprestimo.merge(database_exemplares)
df_final.head(50)
```

## Manipulando os dados

### Adicionando uma nova conluna que coloca a classificação do número da coluna "localização"

 - 000 a 099: Generalidades. Ciência e conhecimento.
 - 100 a 199: Filosofia e psicologia.
 - 200 a 299: Religião.
 - 300 a 399: Ciências sociais.
 - 400 a 499: Classe vaga. Provisoriamente não ocupada.
 - 500 a 599: Matemática e ciências naturais.
 - 600 a 699: Ciências aplicadas.
 - 700 a 799: Belas artes.
 - 800 a 899: Linguagem. Língua. Linguística.
 - 900 a 999: Geografia. Biografia. História."

```python
def CDU_code(code):
  classification = ''
  if code < 99:
    classification = 'Generalidades. Ciência e conhecimento.'
    return classification
  elif code >= 100 and code <199:
    classification = 'Filosofia e psicologia.'
    return classification
  elif code >= 200 and code <299:
    classification = 'Religião.'
    return classification
  elif code >= 300 and code <399:
    classification = 'Ciências sociais.'
    return classification
  elif code >= 400 and code <499:
    classification = 'Classe vaga. Provisoriamente não ocupada.'
    return classification
  elif code >= 500 and code <599:
    classification = 'Matemática e ciências naturais.'
    return classification
  elif code >= 600 and code <699:
    classification = 'Ciências aplicadas.'
    return classification
  elif code >= 700 and code <799:
    classification = 'Belas artes.'
    return classification
  elif code >= 800 and code <899:
    classification = 'Linguagem. Língua. Linguística.'
    return classification
  elif code >= 900 and code <999:
    classification = 'Geografia. Biografia. História.'
    return classification
  else:
    return 'NaN'

lista_classes = []

for cell in df_final['localizacao']:
  lista_classes.append(CDU_code(cell))

df_final['classe_geral'] = lista_classes

df_final.head()


```

### Excluindo a coluna "registro_sistema" por não ter relevancia na análise

```python
df_final = df_final.drop('registro_sistema', axis=1)
df_final.head()
```

### Modificando o dtype da coluna "matricula_ou_siape" para string, a fim de ter uma melhor visualização

```python
df_final['matricula_ou_siape'] = df_final['matricula_ou_siape'].astype('string')
df_final.head()
```
