# DIA 1

## Importando as bases de dados dos emprestimos e concatenando todas.

```python
import pandas as pd

base_url = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/'

files = ['emprestimos-20101.csv',
         'emprestimos-20102.csv',
         'emprestimos-20111.csv',
         'emprestimos-20112.csv',
         'emprestimos-20121.csv',
         'emprestimos-20122.csv',
         'emprestimos-20131.csv',
         'emprestimos-20132.csv',
         'emprestimos-20141.csv',
         'emprestimos-20142.csv',
         'emprestimos-20151.csv',
         'emprestimos-20152.csv',
         'emprestimos-20161.csv',
         'emprestimos-20162.csv',
         'emprestimos-20171.csv',
         'emprestimos-20172.csv',
         'emprestimos-20181.csv',
         'emprestimos-20182.csv',
         'emprestimos-20191.csv',
         'emprestimos-20192.csv',
         'emprestimos-20201.csv']


url_file_emprestimo = [base_url + file for file in files]

dfs_emprestimos = []

for url in url_file_emprestimo:
    df = pd.read_csv(url)
    dfs_emprestimos.append(df)

df_combinado_emprestimo = pd.concat(dfs_emprestimos, ignore_index=True)
```

## Verificando se há duplicatas (Informações duplicadas)

```python
df_combinado_emprestimo.value_counts()
```

### Removendo-as:

```python
df_combinado_emprestimo = df_combinado_emprestimo.drop_duplicates()

display(df_combinado_emprestimo.value_counts())
```

## Importando a segunda base de dados (Exemplares)

```python
database_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
display(database_exemplares)
```

## Mesclando as duas bases de dados

```python
df_final = df_combinado_emprestimo.merge(database_exemplares)
df_final.head(50)
```

# DIA 2

## Manipulando os dados

### Adicionando uma nova conluna que coloca a classificação do número da coluna "localização"

 - 000 a 099: Generalidades. Ciência e conhecimento.
 - 100 a 199: Filosofia e psicologia.
 - 200 a 299: Religião.
 - 300 a 399: Ciências sociais.
 - 400 a 499: Classe vaga. Provisoriamente não ocupada.
 - 500 a 599: Matemática e ciências naturais.
 - 600 a 699: Ciências aplicadas.
 - 700 a 799: Belas artes.
 - 800 a 899: Linguagem. Língua. Linguística.
 - 900 a 999: Geografia. Biografia. História."

```python
def CDU_code(code):
  classification = ''
  if code < 99:
    classification = 'Generalidades. Ciência e conhecimento.'
    return classification
  elif code >= 100 and code <199:
    classification = 'Filosofia e psicologia.'
    return classification
  elif code >= 200 and code <299:
    classification = 'Religião.'
    return classification
  elif code >= 300 and code <399:
    classification = 'Ciências sociais.'
    return classification
  elif code >= 400 and code <499:
    classification = 'Classe vaga. Provisoriamente não ocupada.'
    return classification
  elif code >= 500 and code <599:
    classification = 'Matemática e ciências naturais.'
    return classification
  elif code >= 600 and code <699:
    classification = 'Ciências aplicadas.'
    return classification
  elif code >= 700 and code <799:
    classification = 'Belas artes.'
    return classification
  elif code >= 800 and code <899:
    classification = 'Linguagem. Língua. Linguística.'
    return classification
  elif code >= 900 and code <999:
    classification = 'Geografia. Biografia. História.'
    return classification
  else:
    return None

lista_classes = []

for cell in df_final['localizacao']:
  lista_classes.append(CDU_code(cell))

df_final['classe_geral'] = lista_classes

df_final.head()


```

### Excluindo a coluna "registro_sistema" por não ter relevancia na análise

```python
df_final = df_final.drop('registro_sistema', axis=1)
df_final.head()
```

### Modificando o dtype da coluna "matricula_ou_siape" para string, a fim de ter uma melhor visualização

```python
df_final['matricula_ou_siape'] = df_final['matricula_ou_siape'].astype('string')
df_final.head()
```

## Criação do dataframe que terá a quantidade de emprestimos de cada mês nos anos de 2010-2020

```python
df_alunos_graduacao_acervo_circulante = pd.DataFrame(df_alunos_graduacao.query('colecao == "Acervo Circulante"'))
df_alunos_graduacao_acervo_circulante['data_emprestimo'] = pd.to_datetime(df_alunos_graduacao_acervo_circulante['data_emprestimo'])
df_alunos_graduacao_acervo_circulante['ano'] = df_alunos_graduacao_acervo_circulante['data_emprestimo'].dt.year
df_alunos_graduacao_acervo_circulante['mes'] = df_alunos_graduacao_acervo_circulante['data_emprestimo'].dt.month
df_alunos_graduacao_acervo_circulante = df_alunos_graduacao_acervo_circulante.loc[:, ['ano', 'mes']]
df_alunos_graduacao_acervo_circulante = df_alunos_graduacao_acervo_circulante.value_counts().to_frame('quantidade').reset_index()
df_alunos_graduacao_acervo_circulante
```

## Criação do grafico da quantidade de emprestimos de cada mês nos anos de 2010-2020

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def gera_box_plot(dataset,x,y,titulo,subtitulo):
  '''
  Esta função irá gerar um gráfico de boxplot.

  Dataset = conjunto de dados do gráfico
  x = valor do eixo x do gráfico
  y = valor do eixo y do gráfico
  título = título do gráfico
  subtitulo = subtitulo do texto
  '''

  sns.set_theme(style="darkgrid", palette='Blues',font_scale=1.3)
  plt.figure(figsize=(16,10))

  ax = sns.boxplot(y= y, x= x, data= dataset,color='#4171EF')
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))


  plt.ylim(0,max(dataset[y])*1.1)                                               #Definir o limite do eixo y
  plt.xlabel(None)
  plt.ylabel(None)

  ax.set_title(titulo+"\n",size=20,loc='left',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset[y])*1.11,fontsize=18, ha='left',color='gray')
```

```python
gera_box_plot(df_alunos_graduacao_acervo_circulante,'ano','quantidade','Distribuição dos empréstimos mensais', 'Realizados pelos alunos de graduação no acervo circulante')
```

## Análise do boxplot:
 1. Levando em conta que o vinculo de aluno que mais realiza emprestiimos na biblioteca é o de graduação, e que por sua vez tem como coleção mais requisitada a do Acervo Circulante, este boxplot pode ser visto como grande parte do conjunto de dados fornecidos.

 2. Nos anos de 2010-2013 houveram um aumento significativo de emprestimos, tendo um declinio em 2014, porém um mais acentuado em 2019-2020, provavelmente por conta da pandemia.

 3. De 2014-2017 houve um "equilibrio" nos emprestimos, onde o q3 se manteve quase na mesma linha durante esse período.

 4. Dessa forma, cabe uma análise profunda das estrategias ed negocio tomadas em 2010-2013 para identificar o que ajudou a ter um aumento constante e, em contrapartida, identificar o que houve a partir de 2014. Feito isso, a direção pode tomar estrategias de 2010-2013, adaptadas para os tempos atuais, e tambem rever as estrategias tomadas em 2014 que geraram o declino, afim de não se repetir mais.
